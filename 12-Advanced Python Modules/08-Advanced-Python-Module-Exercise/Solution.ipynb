```python
import shutil
import os
import re

extract_filename = r"C:\\Users\\User\\Documents\\Work\\Complete-Python-3-Bootcamp-2025\\12-Advanced Python Modules\\08-Advanced-Python-Module-Exercise\\unzip_me_for_instructions.zip"
dir_to_unzip = 'C:\\Users\\User\\Documents\\Work\\Complete-Python-3-Bootcamp-2025\\12-Advanced Python Modules\\08-Advanced-Python-Module-Exercise\\unzip_me' # Change to your desired path
shutil.unpack_archive(extract_filename,dir_to_unzip,'zip')
phonetext = '###-###-####'

for folder , sub_folders , files in os.walk("unzip_me"):
     
    print("Currently looking at folder: "+ folder)
    print('\n')
    print("THE SUBFOLDERS ARE: ")
    for sub_fold in sub_folders:
        print("\t Subfolder: "+sub_fold )
    
    print('\n')
    
    print("THE FILES ARE: ")
    for f in files:
        print("\t File: "+f)
        
        file_path = os.path.join(folder, f)

        # Try to open and read the file
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                contents = file.read()
                pattern = r'\d{3}-\d{3}-\d{4}'
                matches = re.findall(pattern, contents)
                print(matches)                   
                
        except Exception as e:
            print(f"\t[Could not read file: {e}]\n")
        
    print('\n')
    
    
    
```
