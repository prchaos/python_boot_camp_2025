```python
res = requests.get('https://books.toscrape.com/')

soup = bs4.BeautifulSoup(res.text,"lxml")

category_links = soup.select(".side_categories ul ul li a")

# Extract the text from each <a> and clean it up
categories = [link.get_text(strip=True) for link in category_links]

print(categories)
```

```python
for item in soup.select(".side_categories ul ul li a"):
    print(item.text)
```

```python
import requests
import bs4
res = requests.get('https://books.toscrape.com/')
```

```python
soup = bs4.BeautifulSoup(res.text, "lxml")
```

```python
soup.select('.thumbnail')
```

```python
for item in soup.select(".thumbnail"):
    print(item)
```

```python
for item in soup.select(".thumbnail"):
    print(item[4])
    print(item[6])
```

```python
# Get the list of all elements first
items = soup.select(".thumbnail")

# Print only the 5th and 7th items (index 4 and 6)
if len(items) >= 7:
    print(items[4])
    print(items[7])
else:
    print("Not enough images found.")
```

```python
items['src'][0]
```

```python
items = soup.select(".thumbnail")

# Make sure there's at least one item
if items:
    print(items[3]['alt'])
    print(items[9]['src'])
else:
    print("No images found.")
```
